```python
data['published_at'] = pd.to_datetime(data['published_at'], utc=True)
data['year'] = data['published_at'].dt.year
vacancies_by_year = data['year'].value_counts().sort_index()
print('Вакансии за 2012 год:', vacancies_by_year.get(2012, 0))
print('Вакансии за 2013 год:', vacancies_by_year.get(2013, 0))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(vacancies_by_year.index, vacancies_by_year.values)
plt.xlabel('Год')
plt.ylabel('Количество вакансий')
plt.title('Динамика количества вакансий по годам')
plt.xticks(vacancies_by_year.index.astype(int))
plt.show()
```
